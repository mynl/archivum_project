```python
# other ways to find orphans are all the same
orphan_docs1 = (sorted(list(set(b.doc_df.path) - set(b.ref_doc_df.path))))
len(orphan_docs1)

vc = b.ref_doc_df.groupby('path').count()

orphan_docs2 =  (sorted(list(set(b.doc_df.path) - set(vc.index))))
len(orphan_docs2)

orphan_docs3 = set(b.doc_df.path) - set(b.ref_doc_df.path)
len(orphan_docs3)

od = b.docs_no_refs()
assert np.all((
    sorted(orphan_docs1) == sorted(orphan_docs2), 
    sorted(orphan_docs1) == sorted(orphan_docs3), 
    sorted(orphan_docs2) == sorted(orphan_docs3),
    sorted(od.path) == sorted(orphan_docs1))
)
```
