```python
import fitz  # pymupdf

def extract_pdf_metadata(path: Path) -> dict:
    with fitz.open(path) as doc:
        return doc.metadata
```

```python
extract_pdf_metadata(bit[0])
```

```python
df = pd.DataFrame([extract_pdf_metadata(f) for f in sample(ps, 100) ] )
```
